## Mini Project 1

### FizzBuzz Game

```python
import random

print(".FizzBuzz Game.")
last= 0
score= 0
for num in range(1, 21):
    current = random.randint(1, 30)
    print(f"Computer Number: {current}")
    total = last + current

    if total % 3== 0 and total % 5== 0:
        correct = "fizzbuzz"
    elif total % 3== 0:
        correct = "fizz"
    elif total % 5== 0:
        correct = "buzz"
    else:
        correct = str(total)

    d = input("Your guess: ")
    if d == correct:
        print("Correct")
        score += 1
        print(f"(Score is: {score})")
    else:
        print("Wrong")
        print(f"Correct answer is: {correct}")
        print("Game Over")
        break

    last = total
print(f"Final Score is: {score}")

```

### Movie Budget Analysis

```python
movies = [
    ("Eternal Sunshine of the Spotless Mind", 20000000),
    ("Memento", 9000000),
    ("Requiem for a Dream", 4500000),
    ("Pirates of the Caribbean: On Stranger Tides", 379000000),
    ("Avengers: Age of Ultron", 365000000),
    ("Avengers: Endgame", 356000000),
    ("Incredibles 2", 200000000)
]

user_movies = []

def add_movie():
    print(".Add a New Movie.")
    name = input("Enter movie name:")
    try:
        budget = int(input("Enter movie budget: "))
        user_movies.append((name, budget))
        print(f"Movie '{name}' added.")
    except ValueError:
        print("Please enter a budget")

def show_average():
    print("..Movie Budget Analysis..")
    data = user_movies if user_movies else movies
    avg = sum(b for _, b in data) / len(data)
    print(f"Average movie budget: ${avg:.2f}\n")

    above = [(n, b - avg) for n, b in data if b > avg]

    if above:
        print("Movies above average budget:")
        for name, gap in above:
            print(f"- {name} (Above by ${gap:.2f})")
        print(f"\nTotal movies above average: {len(above)}\n")
    else:
        print("No movies have a higher budget than the average.\n")

while True:
    print("==== Movie Budget Menu ====")
    print("1. Add a movie")
    print("2. Show budget info")
    print("3. Exit")

    choice = input("Enter your choice (1-3): ").strip()

    if choice == "1":
        add_movie()
    elif choice == "2":
        show_average()
    elif choice == "3":
        print("Exit")
        break
    else:
        print("Invalid choice")
```
