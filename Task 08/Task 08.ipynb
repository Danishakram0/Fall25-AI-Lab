```python
def minmax(depth,node_index,is_max,scores,target_depth):
    if depth == targ_depth:
        return scores[node_idx]
    
    if is_max:
        return max(minmax(depth+1,node_idx*2,False,scores,targ_depth),
                   minmax(depth+1,node_idx*2+1,False,scores,targ_depth))
    
    else:
        return min(minmax(depth+1,node_idx*2,True,scores,targ_depth),
                   minmax(depth+1,node_idx*2+1,True,scores,targ_depth))
    

scores = [7,3,9,1,0,6,5,4]
depth = 0
node_idx = 0
is_max = True
targ_depth = 3
opt_value = minmax(depth,node_idx,is_max,scores,targ_depth)

print("The Ideal value is :",opt_value)
    
```

```python
def minimax(depth, index, is_max_turn, scores, max_depth):
    if depth == max_depth:
        return scores[index]

    if is_max_turn:
        left = minimax(depth + 1, index * 2, False, scores, max_depth)
        right = minimax(depth + 1, index * 2 + 1, False, scores, max_depth)
        return max(left, right)
    else:
        left = minimax(depth + 1, index * 2, True, scores, max_depth)
        right = minimax(depth + 1, index * 2 + 1, True, scores, max_depth)
        return min(left, right)

scores = [7, 3, 9, 1, 0, 6, 5, 4]

starting_depth = 0
starting_index = 0
is_max_turn = True  
max_tree_depth = 3    
best_value = minimax(starting_depth, starting_index, is_max_turn, scores, max_tree_depth)

print("The optimal value is:", best_value)
```
