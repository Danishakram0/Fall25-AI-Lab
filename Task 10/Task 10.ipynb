<div class="alert alert-block alert-warning">
    4.1. <b>Manage Null Values</b>
</div>

```python
# Check if 'BMI' exists before dropping
if 'BMI' in df.columns:
    df = df.drop('BMI', axis=1)
```

```python
# Check if 'Age' exists before dropping
if 'Age' in df.columns:
    df = df.drop('Age', axis=1)
```

```python
# For Object / String Type
def fillnaObjectMode(cols):
    for i in cols:
        if i in df.columns:
            df[i] = df[i].fillna(df[i].mode()[0])

columns = [col for col in ['Age', 'Glucose', 'BloodPressure'] if col in df.columns]
```

```python
# For Int Type
def fillnaIntMode(cols):
    for i in cols:
        if i in df.columns:
            df[i] = df[i].fillna(df[i].mode()[0])

columns = [col for col in ['Glucose','BloodPressure','Heart Disease','Chest pain type','Slope of ST','ST depression'] if col in df.columns]
fillnaIntMode(columns)
```

```python
# For Float Type
def fillnaFloatMean(cols):
    for i in cols:
        if i in df.columns:
            df[i] = df[i].fillna(df[i].mean())

columns = [col for col in ['Age','Cholesterol','BP','Heart Disease','Chest pain type','Slope of ST','ST depression'] if col in df.columns]
fillnaFloatMean(columns)
```

```python
# Safely convert 'Age' to int if it exists
if 'Age' in df.columns:
    df['Age'] = df['Age'].astype('int64')
# df['price'].astype('int64', inplace=True)
```

```python
# For Float Type
def changeFloatToInt(cols):
    for i in cols:
        if i in df.columns:
            df[i] = df[i].astype('int64')

columns = [col for col in ['Glucose','BloodPressure','BMI','Insulin','Pregnancies','SkinThickness'] if col in df.columns]
changeFloatToInt(columns)
```

```python
def convertObjToInt(cols):
    for i in cols:
        if i in df.columns:
            data = pd.DataFrame({i:df[i].unique()})
            data_LE = LabelEncoder()
            data_LE.fit(np.ravel(data))    # data_LE.fit(data)
            df[i] = data_LE.transform(df[i])
        
columns = [col for col in ['Glucose','BloodPressure','BMI','Insulin','Pregnancies','SkinThickness'] if col in df.columns]
convertObjToInt(columns)
```
