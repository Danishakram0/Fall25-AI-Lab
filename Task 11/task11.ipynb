Data Pre-Processing

Manage Null Values

```python

df = df.drop('Age', axis=1)
```

```python

df = df.drop('BMI', axis=1)
```

```python

def fillnaObjectMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns =[ 'Glucose', 'BloodPressure', 'Insulin', 'DiabetesPedigreeFunction', 'Outcome']
fillnaObjectMode(columns)
```

```python

def fillnaIntMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns =[ 'Glucose', 'BloodPressure', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
fillnaIntMode(columns)
```

```python

def fillnaFloatMean(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mean())


columns =[ 'Glucose', 'BloodPressure', 'Insulin', 'DiabetesPedigreeFunction', 'Outcome']
fillnaFloatMean(columns)
```

Applying Classifier / Model Training

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
def convertObjToInt(cols):
    for i in cols:
        label = LabelEncoder()
        df[i] = label.fit_transform(df[i])
        
columns = [ 'Outcome', 'Glucose', 'BloodPressure', 'Insulin',]
convertObjToInt(columns)
```

```python
X = df.drop("Diabetes", axis=1)
y = df["Diabetes"]
```

```python
train_X, test_X, train_y, test_y = train_test_split(
    X, y, test_size=0.3, shuffle=True, random_state=42
)
```

```python
model_svc = SVC()

model_svc.fit(train_X, train_y)

print("Model trained successfully!")
```

```python
accuracy = model_svc.score(test_X, test_y)
print("Accuracy:", accuracy)
```

```python
import pickle
pickle.dump(model_svc, open('model_svc.pkl', 'wb'))   
```
